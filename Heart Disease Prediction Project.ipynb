```python
# Data Modeling
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(features,target,test_size=0.25,random_state=40)
```

```python
xtrain.shape,xtest.shape,ytrain.shape,ytest.shape
```

```python
# Algorithm Implementation
```

```python
# Decision Tree Classifier
```

```python
from sklearn.tree import DecisionTreeClassifier
model= DecisionTreeClassifier(max_depth=5,criterion="entropy")
model.fit(xtrain,ytrain)
ypred=model.predict(xtest)

from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
a=accuracy_score(ytest,ypred)
cm=confusion_matrix(ytest,ypred)
cr=classification_report(ytest,ypred)
print(a)
print(cm)
print(cr)
```

```python
# we get the accuracy of the Decision Tree classifier is about 83%.
```

```python
# Random Forest Classifier
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=100,max_depth=5)
model.fit(xtrain,ytrain)
ypred=model.predict(xtest)

from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
a=accuracy_score(ytest,ypred)
cm=confusion_matrix(ytest,ypred)
cr=classification_report(ytest,ypred)
print(a)
print(cm)
print(cr)
```

```python
#  we can get the accuracy of the Random Forest classifier is about 89%.
```
